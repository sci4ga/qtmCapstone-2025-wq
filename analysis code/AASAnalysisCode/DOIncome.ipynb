```python
import pandas as pd
import numpy as np
```

```python
doData = pd.read_csv('/Users/keerthyrangan/Downloads/chemicalDataCensusTractCleaned2.csv')
incomeData = pd.read_csv('/Users/keerthyrangan/Downloads/Income Percentile(in) (2).csv')
```

```python

# Merge on tract_id
merged_df = pd.merge(doData, incomeData, left_on='Census Tract', right_on='FPS_ID', how='outer')

```

```python
merged_df
```

```python
merged_df = merged_df.dropna()
```

```python
merged_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Melt the DataFrame to long-form
df_melted = merged_df.melt(id_vars='Dissolved Oxygen Threat levels', value_vars=['LOWINCPCT'],
                    var_name='Type', value_name='Value')

# Create the barplot
sns.barplot(data=df_melted, x='Dissolved Oxygen Threat levels', y='Value', hue='Type')
plt.title('Low Income Percentages by Dissolved Oxygen Threat levels')
plt.ylabel('Low Income Percentages')
plt.show()

```

```python
# Create a list of arrays, one per category
groups = [group['LOWINCPCT'].values for name, group in merged_df.groupby('Dissolved Oxygen Threat levels')]
```

```python
from scipy.stats import f_oneway

f_stat, p_val = f_oneway(*groups)
print(f"F-statistic: {f_stat:.4f}, P-value: {p_val:.4e}")
```
