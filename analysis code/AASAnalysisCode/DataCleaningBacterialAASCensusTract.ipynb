```python
import pandas as pd
import numpy as np
```

Reading Datasets 

```python
bacterialData = pd.read_csv("/Users/keerthyrangan/Downloads/bacteria(Sheet1).csv", encoding = "latin-1")
bacterialData
```

Bacterial

```python
bacterialData["EventDate"] = pd.to_datetime(bacterialData["EventDate"])

bacterialData = bacterialData[bacterialData["EventDate"].dt.year == 2022]
bacterialData
```

```python
bacterialData = bacterialData[bacterialData["min_temp"] > 10]
bacterialData = bacterialData[bacterialData["min_temp"] < 49]
bacterialData = bacterialData[bacterialData["plate_blank"] == 0]

bacterialData
```

```python
variables_to_keep = ["EventDate","SiteID","SiteName","site_latitude","site_longitude","plate_blank","plate_one",
                     "plate_two","plate_three","colony_avg","min_temp","max_temp"]
bacterialData = bacterialData[variables_to_keep]
bacterialData
```

```python
bacterialData = bacterialData.dropna()
bacterialData
```

```python
bacterialData = bacterialData[bacterialData["plate_one"] != "TNTC"]
bacterialData = bacterialData[(bacterialData["plate_two"] != "TNTC") & (bacterialData["plate_two"] != " ")]
bacterialData = bacterialData[bacterialData["plate_three"] != "TNTC"]

```

```python
bacterialData["plate_one"] = bacterialData["plate_one"].astype(float)
bacterialData["plate_two"] = bacterialData["plate_two"].astype(float)
bacterialData["plate_three"] = bacterialData["plate_three"].astype(float)
bacterialData["colony_avg"] = bacterialData["colony_avg"].astype(float)
```

```python
bacterialData["plate_one"] = bacterialData["plate_one"] * 100
bacterialData["plate_two"] = bacterialData["plate_two"] * 100
bacterialData["plate_three"] = bacterialData["plate_three"] * 100
```

```python
bacterialData["1st Single Sample E Coli Threat Level"]= pd.cut(bacterialData['plate_one'], 
                        bins=[-1, 100, 235,max(bacterialData['plate_one'])],  # Define numeric ranges
                        labels=['Water Very Clean', 'Still in Acceptable Threshold', 'High Risk for Waterborne Illness']) 

bacterialData["2nd Single Sample E Coli Threat Level"]= pd.cut(bacterialData['plate_two'], 
                        bins=[-1, 100, 235,max(bacterialData['plate_two'])],  # Define numeric ranges
                        labels=['Water Very Clean', 'Still in Acceptable Threshold', 'High Risk for Waterborne Illness']) 

bacterialData["3rd Single Sample E Coli Threat Level"]= pd.cut(bacterialData['plate_three'], 
                        bins=[-1, 100, 235,max(bacterialData['plate_three'])],  # Define numeric ranges
                        labels=['Water Very Clean', 'Still in Acceptable Threshold', 'High Risk for Waterborne Illness']) 

bacterialData["E Coli Colony Average Threat Level"] = pd.cut(bacterialData['colony_avg'], 
                        bins=[-1, 100, 410,max(bacterialData['colony_avg'])],  # Define numeric ranges
                        labels=['Rarely Unsafe', 'Sometimes Unsafe', 'Often Unsafe']) 
```

```python
bacterialData
```

```python

bacterialData = bacterialData.rename(columns={"plate_one": "1st Sample E Coli Colony Count","plate_two":"2nd Sample E Coli Colony Count",
                              "plate_three":"3rd Sample E Coli Colony Count",
                              "colony_avg":"E Coli Colony Average"})
bacterialData = bacterialData.drop(["plate_blank","min_temp","max_temp"],axis=1)
bacterialData
```

```python
from censusgeocode import CensusGeocode

cg = CensusGeocode()

lat = 33.7490
lon = -84.3880
def get_tract_from_coords(lat, lon):
    try:
        result = cg.coordinates(x=lon, y=lat)
        return result['Census Tracts'][0]['GEOID']
    except Exception:
        return None  # or 'NA' if you prefer

get_tract_from_coords(lat, lon)
```

```python
# Apply function
bacterialData['Census Tract'] = bacterialData.apply(
    lambda row: get_tract_from_coords(row['site_latitude'], row['site_longitude']),
    axis=1
)

print(bacterialData)
```

```python
bacterialData = bacterialData.drop(["site_latitude","site_longitude"],axis=1)
```

```python
bacterialData
```

```python
bacterialData = bacterialData.groupby("Census Tract").agg({
    "SiteID": "min",
    "EventDate": "mean",
    "E Coli Colony Average": lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan,
    "E Coli Colony Average Threat Level": lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan,
}).reset_index()


bacterialData
```
