```python
import pandas as pd
import numpy as np
```

```python
chemicalData = pd.read_csv("/Users/keerthyrangan/Downloads/chemical.csv", encoding = "latin-1")
chemicalData
```

```python
chemicalData["EventDate"] = pd.to_datetime(chemicalData["EventDate"])

chemicalData = chemicalData[chemicalData["EventDate"].dt.year == 2022]
chemicalData
```

```python
chemicalData = chemicalData[chemicalData["water_temp"] < 20]
chemicalData = chemicalData[(chemicalData["ph1"] >= 6.5) & (chemicalData["ph1"] <= 8.5)]

chemicalData
```

```python
variables_to_keep = ["EventDate","SiteID","SiteName","site_latitude","site_longitude","water_temp","ph1","ph2",
                     "do_saturation","DissolvedOxygen1","DissolvedOxygen2"]
chemicalData = chemicalData[variables_to_keep]
chemicalData
```

```python
chemicalData = chemicalData.dropna()
chemicalData
```

```python
chemicalData.dtypes
```

```python
chemicalData["pH_avg_at_site"] = (chemicalData["ph1"] + chemicalData["ph2"])/2
chemicalData["DO_avg_at_site"] = (chemicalData["DissolvedOxygen1"] + chemicalData["DissolvedOxygen2"])/2
chemicalData
```

```python
chemicalData["Dissolved Oxygen Threat Levels"] = pd.cut(chemicalData['DO_avg_at_site'], 
    bins=[min(chemicalData['DO_avg_at_site']), 3, 6.5, max(chemicalData['DO_avg_at_site'])], 
    labels=['Often Unsafe DO Levels', 'Sometimes Unsafe DO levels', 'Rarely Unsafe DO Levels']
)

chemicalData
```

```python
import numpy as np

chemicalDataSites = chemicalData.groupby("SiteName").agg({
    "SiteID": "min",
    "EventDate": "mean",
    "DO_avg_at_site": lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan,
    "Dissolved Oxygen Threat Levels": lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan,
    "site_latitude": "min",
    "site_longitude": "min"
}).reset_index()
```
