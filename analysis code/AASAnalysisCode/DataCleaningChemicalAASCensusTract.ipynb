```python
import pandas as pd
import numpy as np
```

```python
chemicalData = pd.read_csv("/Users/keerthyrangan/Downloads/chemical.csv", encoding = "latin-1")
chemicalData
```

```python
chemicalData["EventDate"] = pd.to_datetime(chemicalData["EventDate"])

chemicalData = chemicalData[chemicalData["EventDate"].dt.year == 2022]
chemicalData
```

```python
chemicalData = chemicalData[chemicalData["water_temp"] < 20]
chemicalData = chemicalData[(chemicalData["ph1"] >= 6.5) & (chemicalData["ph1"] <= 8.5)]

chemicalData
```

```python
variables_to_keep = ["EventDate","SiteID","SiteName","site_latitude","site_longitude","water_temp","ph1","ph2",
                     "do_saturation","DissolvedOxygen1","DissolvedOxygen2"]
chemicalData = chemicalData[variables_to_keep]
chemicalData
```

```python
chemicalData = chemicalData.dropna()
chemicalData
```

```python
chemicalData["pH_avg_at_site"] = (chemicalData["ph1"] + chemicalData["ph2"])/2
chemicalData["DO_avg_at_site"] = (chemicalData["DissolvedOxygen1"] + chemicalData["DissolvedOxygen2"])/2
chemicalData
```

```python
import statsmodels.api as sm

# Add a constant to the independent variable
X = sm.add_constant(chemicalData['water_temp'])  # adds intercept term
y = chemicalData['DO_avg_at_site']

# Fit the regression model
model = sm.OLS(y, X).fit()

# View the summary
print(model.summary())

DO_pred = model.predict(X)

# Optional: Add to DataFrame
chemicalData['DO_predicted'] = DO_pred
```

```python
conditions = [
    (chemicalData['DO_avg_at_site'] < chemicalData['DO_predicted']),
    (chemicalData['DO_avg_at_site'] >= chemicalData['DO_predicted'])
]

choices = ["Unsafe","Safe"]

chemicalData['Dissolved Oxygen Threat levels'] = np.select(conditions, choices)


chemicalData
```

```python
from censusgeocode import CensusGeocode

cg = CensusGeocode()

lat = 33.7490
lon = -84.3880
def get_tract_from_coords(lat, lon):
    try:
        result = cg.coordinates(x=lon, y=lat)
        return result['Census Tracts'][0]['GEOID']
    except Exception:
        return None  # or 'NA' if you prefer

get_tract_from_coords(lat, lon)
```

```python
# Apply function
chemicalData['Census Tract'] = chemicalData.apply(
    lambda row: get_tract_from_coords(row['site_latitude'], row['site_longitude']),
    axis=1
)

print(chemicalData)
```

```python
def dissolved_oxygen(series):
    safe_prop = series.value_counts(normalize=True).get('Safe', 0)
    
    if safe_prop <= 0.05:
        return 'Often Unsafe'
    elif safe_prop <= 0.5:
        return 'Tends to be Unsafe'
    elif safe_prop <= 0.8:
        return 'Sometimes Unsafe'
    else:
        return 'Rarely Unsafe'
```

```python
chemicalDataSites = chemicalData.groupby("Census Tract").agg({
    "SiteID": "min",
    "EventDate": "mean",
    "DO_avg_at_site": lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan,
    "Dissolved Oxygen Threat levels": dissolved_oxygen,
    "site_latitude": "min",
    "site_longitude": "min"
}).reset_index()
```
