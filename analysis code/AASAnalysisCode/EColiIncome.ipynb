```python
variables_to_keep = ["EventDate","SiteID","Census Tract","E Coli Colony Average","E Coli Colony Average Threat Level","LOWINCPCT",
                     "Population density (people per square mile), 2022"]
merged_df2 = merged_df2[variables_to_keep]
merged_df2
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Melt the DataFrame to long-form
df_melted = urban_mergeddf2.melt(id_vars='E Coli Colony Average Threat Level', value_vars=['LOWINCPCT'],
                    var_name='Type', value_name='Value')

# Create the barplot
sns.barplot(data=df_melted, x='E Coli Colony Average Threat Level', y='Value', hue='Type')
plt.title('Low Income Percentages by E Coli Colony Average Threat Levels (Urban Areas)') 
plt.ylabel('Low Income Percentages')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Melt the DataFrame to long-form
df_melted = rural_mergeddf2.melt(id_vars='E Coli Colony Average Threat Level', value_vars=['LOWINCPCT'],
                    var_name='Type', value_name='Value')

# Create the barplot
sns.barplot(data=df_melted, x='E Coli Colony Average Threat Level', y='Value', hue='Type')
plt.title('Low Income Percentages by E Coli Colony Average Threat Levels (Rural Areas)')
plt.ylabel('Low Income Percentages')
plt.show()
```

```python
# Create a pivot table with counts
pivot = rural_mergeddf2.pivot_table(index='E Coli Colony Average Threat Level', columns='Income Levels', aggfunc='size', fill_value=0)

# Plot heatmap
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title('Heatmap of Counts by Income Level and E Coli Threat Levels (Rural Areas)')
plt.show()

```
